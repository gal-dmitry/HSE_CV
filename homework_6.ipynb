# Поиск изображений по содержанию (30 (+20) баллов)

Вам предстоит построить систему, которая позволит выполнить поиск изображений по содержанию в базе Corel-10k, которая приложена к домашнему заданию. База состоит из 100 классов изображений по 100 изображений в каждом классе. Имя каждого изображения в базе состоит из **X.JPEG**, где X - число от 1 до 10000. Изображения из одного класса лежат подряд в промежутке 1-100, 101-200, и тд. В файле **test.dat** лежит список картинок, которые не должны участвовать в построении поисковой базы, но по которым надо будет сделать 1000 поисковых запросов. Ограничение по памяти -- 2Гб.

1. По изображениям не попавшим в **test.dat** посчитайте дескриптор с помощью алгоритма GIST.
   
2. Разбейте дескрипторы всех изображений на 100 кластеров с помощью K-Means.

3. Для каждого кластера постройте функцию хэширования (LSH), с помощью которой закодируйте каждое изображение в тренировочной выборке.

5. Релизуйте функцию `retrieve` продемонстрируйте работу `retrieve` на 5-ти случайных изображениях из **test.dat**. Замерьте время исполнения поискового запроса. Сделайте возможность вызвать `retrieve` без построения индекса, то есть приложите в решение индексированную базу, если только время построения индекса не укладывается в 1 минуту.  

5. Сделайте 1000 запросов изображениями из файла **test.dat**.  Замерьте среднее время исполнения поискового запроса в colab ноутбуке.

    
6. Для каждого запроса нужно оценить AP@k, где k=10 и посчитать среднее значение этой величины по всем запросам (MAP). Формулу берем из этой презентации: https://web.stanford.edu/class/cs276/handouts/EvaluationNew-handout-1-per.pdf

Так же за это задание можно получить еще до 20-ти дополнительных баллов. 

1. Вы можете сделать быстрый поиск. Если качество вашего поиска окажется выше, чем медиана качества по всем поисковым движкам, то вы принимаете участие в 
борьбе за 10 призовых баллов. Первое место получает 10 баллов, второе 9, третье 8 и т.д. 

2. Вы можете сделать хороший поиск. Если ваш поиск окажется быстрее, чем медиана среднего времени работы других участников, то вы принимаете участие в борьбе за точный поиск. Распределение баллов здесь будет такое же как и в конкурсе на скорость. 

Для достижения лучших результатов вам придется регулировать несколько параметров алгоритма:
- Дескриптор. Можете взять что угодно вместо GIST
- Количество кластеров k-means
- Длину кода LSH
- etc 

Принципиально нельзя использовать готовые поисковые движки (annoy, faiss, etc, ..). И помните о том, что время поиска – очень важный параметр.
В случае использования нейросети инференс картинки производите без GPU ускорителя. Для обучения использовать GPU можно. Безусловно **не используйте изобржения из test.dat для обучения сети или построения поисковой базы**.

### Gist

````
class GIST: 
    def __init__(self):
        self.size = (10, 10)
        self.resize = (128, 128)
        self.sigmas = [2., 4., 6., 8.]
        self.thetas = [(2 * np.pi / 8) * i for i in range(8)]
        self.lmbd = 5.
        self.gamma = 3.
        self.psi = 0.


    def gabor(self, image, sigma, theta):
        kernel = cv2.getGaborKernel(self.size,
                                    sigma, 
                                    theta, 
                                    self.lmbd, 
                                    self.gamma, 
                                    self.psi)
        return cv2.filter2D(image, cv2.CV_8UC3, kernel)


    def features(self, gabor):
        h, w, _ = gabor.shape
        h_step = h // 4 + 1
        w_step = w // 4 + 1
        features = []

        for h_start in range(0, h, h_step):
            for w_start in range(0, w, w_step):

                h_stop = h_start + h_step
                w_stop = w_start + w_step

                region_mean = gabor[h_start:h_stop, w_start:w_stop].mean()
                features.append(region_mean)

        return np.array(features)


    def collection(self, image):
        collection = []
        image = image.copy()
        image = cv2.resize(image, self.resize)

        for sigma in self.sigmas:
            for theta in self.thetas:
                gabor = self.gabor(image, sigma, theta)
                features = self.features(gabor)
                collection.append(features)

        return np.concatenate(collection)


    def collection_all(self, images):
        return np.array([self.collection(image) for image in images])
````

````
def get_histogram(name_of_stat, stat):

    x_min = round(np.min(stat)) - 1
    x_max = round(np.max(stat)) + 1
    count = x_max - x_min + 1
    n_bins = np.linspace(x_min, x_max, count)
    
    fig, ax = plt.subplots()
    fig.suptitle('{0}'.format(name_of_stat), fontsize=14)

    ax.hist(stat, bins=n_bins, alpha=0.5, edgecolor='black')
    ax.set_xlabel('{0}'.format(name_of_stat))
    ax.set_ylabel('Count of images')
    
    plt.show()
````

### LSH

````
class LSH:
    def __init__(self, centers, bounds, length):
        self.centers = centers
        self.bounds = bounds
        self.length = length
        self.dct = {}


    def predict_cls(self, point):
        return np.argsort(np.linalg.norm(self.centers - point, axis=1))[0]


    def get_nums(self, cluster_id):
        center = self.centers[cluster_id]
        size = len(center)
        low, high = self.bounds

        nums = np.zeros((self.length, size + 1))
        nums[:, :-1] = np.random.uniform(low, high, (self.length, size))
        nums[:, -1] = -(nums[:, :-1] * center).sum(axis=1)
        nums[:, -1] += np.random.uniform(-100, 100, nums.shape[0])

        self.dct[cluster_id] = nums


    def encode(self, point):
        cluster_id = self.predict_cls(point)

        if cluster_id not in self.dct:
            self.get_nums(cluster_id)

        point = np.concatenate((point, np.array([1])))
        nums = self.dct[cluster_id]
        scores = nums @ point
        return scores >= 0


    def get_lsh(self, points):
        return np.array([self.encode(point) for point in points])
````

````
%%time
centers = kmeans.cluster_centers_
bounds = (-1_000., 1_000.)
length = 100

lsh = LSH(centers, bounds, length)
hash = lsh.get_lsh(descriptors)
````

### Retrieve

````
class CBIR:
    def __init__(self, 
                 lsh, 
                 gist,
                 centers,
                 inverted_indx,
                 hash):
        
        self.lsh = lsh
        self.gist = gist
        self.centers = centers
        self.inverted_indx = inverted_indx
        self.hash = hash


    def predict_cls(self, point):
        return np.argsort(np.linalg.norm(self.centers - point, axis=1))[0]


    def retrieve(self, image, top_k=1, return_distances=False):
        
        gist_code = self.gist.collection(image)
        cluster_id = self.predict_cls(gist_code)

        neighbors_indx = self.inverted_indx[cluster_id]
        neighbors_hash = self.hash[neighbors_indx]

        image_hash = self.lsh.encode(gist_code)

        distances = (neighbors_hash != image_hash).sum(axis=1)
        closest_indx = np.argsort(distances)[:top_k]
        indx = neighbors_indx[closest_indx]

        if return_distances:
            return indx, np.sort(distances)[:top_k]

        return indx


    def retrieve_all(self, images, top_k=1, return_distances=False):
        return [self.retrieve(image, top_k, return_distances) for image in images] 
````

````
def get_indx(kmeans):

    inv_indx = defaultdict(list)

    for indx, label in enumerate(kmeans.labels_):
        inv_indx[label].append(indx)

    for cluster_id in range(kmeans.n_clusters):
        inv_indx[cluster_id] = np.array(inv_indx[cluster_id])

    return inv_indx
````

````
inv_indx = get_indx(kmeans)
cbir = CBIR(lsh, gist, centers, inv_indx, hash)
````

### Quality

````
def precision_k(pred, gt, k):
    matches = (pred == gt)
    return sum(matches) / k

def avg_precision_k(pred, gt, k):
    matches = (pred == gt)

    if sum(matches) == 0:
        return 0

    matches_indices = np.argwhere(matches)
    precisions = [precision_k(pred[:indx[0] + 1], gt, indx[0] + 1) for indx in matches_indices]

    return np.mean(precisions)

def name2class(image_name):
    image_order_number = int(image_name.split(".")[0])
    return (image_order_number - 1) // 100
````

````
avg_precisions = []

for test_name, test_image in tqdm(list(zip(test_split, test_images))):
    image_class = name2class(test_name)

    retrieved_indx = cbir.retrieve(test_image, top_k=10)
    retrieved_names = [train_split[index] for index in retrieved_indx]
    retrieved_classes = np.array([name2class(name) for name in retrieved_names])

    avg_precision = avg_precision_k(retrieved_classes, image_class, k=10)
    avg_precisions.append(avg_precision)

print(f"MAP: {np.mean(avg_precisions)}")
````
